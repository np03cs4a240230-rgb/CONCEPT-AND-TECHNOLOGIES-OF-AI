```python
#problem1
import numpy as np

# 1. Initialize an empty array with size 2x2 using np.zeros (more predictable than np.empty)
empty_array = np.zeros((2, 2))
print(empty_array)

# 2. Initialize an all-one array with size 4x2 using np.ones and dtype float
ones_array = np.ones((4, 2))
print(ones_array)

# 3. Create a filled array using broadcasting from np.full_like
template = np.array([[0, 0], [0, 0]])
filled_array = np.full_like(template, fill_value=9)
print(filled_array)

# 4. Use np.shape and np.dtype manually with np.zeros
zeros_like_array = np.zeros(np.shape(template), dtype=template.dtype)
print(zeros_like_array)

# 5. Use np.ones with shape and dtype extracted from another array
ones_like_array = np.ones(template.shape, dtype=template.dtype)
print(ones_like_array)

# 6. Convert list to array and reshape to column vector
new_list = [1, 2, 3, 4]
converted_array = np.array(new_list).reshape(-1, 1)
print(converted_array)
```

```python
#problem3
import numpy as np

# Given arrays
x = np.array([[1, 2], [3, 5]])
y = np.array([[5, 6], [7, 8]])
v = np.array([9, 10])
w = np.array([11, 12])

# 1. Add the two arrays
add = x + y
print(add)

# 2. Subtract the two arrays
sub = x - y
print(sub)

# 3. Multiply the array with any integer (e.g., 3)
mul = x * 3
print(mul)

# 4. Square of each element in array x
square = x ** 2
print(square)

# 5. Dot products
dot_vw = np.dot(v, w)       # v and w are 1D
dot_xv = np.dot(x, v)       # x is 2x2, v is 1D
dot_xy = np.dot(x, y)       # x and y are 2x2
print(dot_vw)
print(dot_xv)
print(dot_xy)

# 6. Concatenate x and y along rows; v and w along columns
concat_xy_row = np.concatenate((x, y), axis=0)
concat_vw_col = np.column_stack((v, w))
print(concat_xy_row)
print(concat_vw_col)

# 7. Concatenate x and v â€” this will raise an error due to shape mismatch
try:
    concat_xv = np.concatenate((x, v), axis=0)
except ValueError as e:
    concat_xv_error = str(e)
    print(concat_xv_error)
```
